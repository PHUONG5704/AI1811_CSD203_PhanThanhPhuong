```python
#1

def sum_recursive(n):
    if n == 1:
        return 1
    else:
        return n + sum_recursive(n - 1)

# Test the function
n = int(input("Enter a positive integer: "))
print("The sum of numbers from 1 to", n, "is:", sum_recursive(n))
```

```python
#2

def find_min_recursive(a, n):
    # Base case: If the array has only one element, return that element
    if n == 1:
        return a[0]
    
    # Recursive case: Find the minimum element in the rest of the array
    min_of_rest = find_min_recursive(a, n - 1)
    
    # Compare the minimum of the rest with the last element
    if min_of_rest < a[n - 1]:
        return min_of_rest
    else:
        return a[n - 1]

# Test the function
arr = [5, 3, 8, 1, 9]
size = len(arr)
print("The minimum element in the array is:", find_min_recursive(arr, size))
```

```python
#3

def find_sum_recursive(a, n):
    if n == 0:  # Base case: if array size is 0, return 0
        return 0
    else:
        return a[n - 1] + find_sum_recursive(a, n - 1)

# Test the function
arr = [5, 3, 8, 1, 9]
size = len(arr)
print("The sum of all elements in the array is:", find_sum_recursive(arr, size))
```

```python
#4

def is_palindrome_recursive(a, n):
    if n <= 1:  # Base case: An array of size 0 or 1 is always a palindrome
        return True
    elif a[0] != a[n - 1]:  # If the first and last elements don't match, it's not a palindrome
        return False
    else:
        # Recursively check if the subarray excluding the first and last elements is a palindrome
        return is_palindrome_recursive(a[1:n - 1], n - 2)

# Test the function
arr = [1, 2, 3, 2, 1]
size = len(arr)
if is_palindrome_recursive(arr, size):
    print("The array is a palindrome.")
else:
    print("The array is not a palindrome.")

```

```python
#5

def binary_search_recursive(arr, target, left, right):
    if left > right:
        return False

    mid = (left + right) // 2

    if arr[mid] == target:
        return True
    elif arr[mid] < target:
        return binary_search_recursive(arr, target, mid + 1, right)
    else:
        return binary_search_recursive(arr, target, left, mid - 1)

# Function to perform binary search
def binary_search(arr, target):
    return binary_search_recursive(arr, target, 0, len(arr) - 1)

# Test the function
arr = [1, 3, 5, 7, 9, 11, 13, 15]
target = 11
if binary_search(arr, target):
    print(f"The target {target} is present in the array.")
else:
    print(f"The target {target} is not present in the array.")
```

```python
#10

def addReciprocals(n):
    if n == 1:
        return 1.0
    else:
        return 1.0 / n + addReciprocals(n - 1)

# Test the function
n = int(input("Enter a positive integer: "))
result = addReciprocals(n)
print("The sum of the first", n, "reciprocals is:", result)

```

```python
#11

def stirling_first(n, k):
    if n == k == 0:
        return 1
    elif n > 0 and k == 0:
        return 0
    elif n == 0 and k > 0:
        return 0
    else:
        return stirling_first(n - 1, k - 1) - n * stirling_first(n - 1, k)

# Test the function
n = int(input("Enter n: "))
k = int(input("Enter k: "))

result = stirling_first(n, k)
print(f"The Stirling number of the first kind s({n},{k}) is:", result)

```

```python
#12

class TreeNode:
    def __init__(self, label):
        self.label = label
        self.left = None
        self.right = None

def tree_height(node):
    if node is None:
        return 0
    else:
        # Recursively calculate the height of the left and right subtrees
        left_height = tree_height(node.left)
        right_height = tree_height(node.right)

        # Return the maximum height among left and right subtrees plus 1 for the current node
        return max(left_height, right_height) + 1

# Test the function
# Create a sample binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Calculate the height of the binary tree
height = tree_height(root)
print("Height of the binary tree is:", height)


```

```python
#13

class TreeNode:
    def __init__(self, label):
        self.label = label
        self.left = None
        self.right = None

def tree_size(node):
    if node is None:
        return 0
    else:
        # Count the current node and recursively count nodes in left and right subtrees
        left_size = tree_size(node.left)
        right_size = tree_size(node.right)
        
        # Return the total size, which is 1 (for the current node) plus the sizes of left and right subtrees
        return 1 + left_size + right_size

# Test the function
# Create a sample binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# Calculate the size of the binary tree
size = tree_size(root)
print("Size of the binary tree is:", size)
```
